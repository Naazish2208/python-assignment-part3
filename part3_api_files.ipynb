**Task 2 — API Integration**

```python
# Step 1 — Fetch and Display Products:

BASE_URL = "https://dummyjson.com/products"

# Fetch 20 products and parse to JSON

try:
  res = requests.get(f"https://dummyjson.com/products?limit=20", timeout=5)
  products = res.json().get("products",[])

# Print formatted table

  header = f"{'ID':<4} | {'Title':<30} | {'Category':<13} | {'Price':>8} | {'Rating'}"
  sep    = f"{'-'*4}-+-{'-'*30}-+-{'-'*13}-+-{'-'*8}-+-{'-'*6}"
  print("\n--- All Products (first 20) ---")
  print(header)
  print(sep)
  for p in products:
    print(f"{p['id']:<4} | {p['title'][:30]:<30} | "
          f"{p['category'][:13]:<13} | "
          f"${p['price']:>7.2f} | {p['rating']}")


except requests.exceptions.ConnectionError as e:
  print(f"Error fetching products: {e}")
```

```python
# Step 2 — Filter and Sort:

filtered = [p for p in products if p.get("rating", 0) >= 4.5]
filtered.sort(key=lambda p: p["price"], reverse=True)

print("Products with Rating ≥ 4.5 (sorted by price desc)\n")

if not filtered:
  print("No products matched the filter.")

for p in filtered:
  print(f"{p['title'][:35]:<35}  ${p['price']:>7.2f}  ⭐ {p['rating']}")
```

```python
# Step 3 — Search by Category:

url = f"{BASE_URL}/category/laptops"
try:
        response = requests.get(url, timeout=5)
        laptops = response.json().get("products", [])
        print("List of Laptops\n")
        for laptop in laptops:
            print(f"{laptop['title'][:40]:<40}  ${laptop['price']:.2f}")
except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
```

```python
# Step 4 — POST Request (Simulated):

url = f"{BASE_URL}/add"
payload = {
        "title": "My Custom Product",
        "price": 999,
        "category": "electronics",
        "description": "A product I created via API",
    }
try:
        response = requests.post(url, json=payload, timeout=5)
        print("POST Response (Simulated Add)\n")
        print(response.json())
except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
```

**Task** **3** **—** **Exception** **Handling**

```python
# Part A — Guarded Calculator:

def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"

print("Part A: Guarded Calculator\n")
test_cases = [(10, 2), (10, 0), ("ten", 2)]
for a, b in test_cases:
    result = safe_divide(a, b)
    print(result)
```

```python
# Part B — Guarded File Reader:

def read_file_safe(filename):

  try:
    with open(filename, "r", encoding="utf-8") as f:
      return f.read()
  except FileNotFoundError:
    print(f"Error: File '{filename}' not found.")
  finally:
    print("\nFile operation attempt complete.")

print("\nReading files:")
print(read_file_safe("python_notes.txt"))
print(read_file_safe("ghost_file.txt"))
```

```python
# Part C — Robust API Calls

try:
        res = requests.get("https://this-host-does-not-exist-xyz.com", timeout=5)
except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
except Exception as e:
        print("Error:", e)
```

```python
# Part D — Input Validation Loop

while True:
        raw = input("\nEnter a product ID to look up (1 to 100), or write 'quit' to exit:").strip()

        if raw.lower() == "quit":
            print("\nExiting product lookup.")
            break

        # Validate integer
        try:
            product_id = int(raw)
        except ValueError:
            print("\n⚠ Invalid input — please enter a whole number.")
            continue

        # Validate range
        if not (1 <= product_id <= 100):
            print("\n⚠ Out of range — please enter a number between 1 and 100.")
            continue

        # Make the API call
        url = f"{BASE_URL}/{product_id}"
        try:
            response = requests.get(url, timeout=5)
        except requests.exceptions.ConnectionError:
            print("\nConnection failed. Please check your internet.")
            continue
        except requests.exceptions.Timeout:
            print("\nRequest timed out. Try again later.")
            continue
        except Exception as e:
            print(f"\nUnexpected error: {e}")
            continue

        if response.status_code == 404:
            print("\nProduct not found.")
        elif response.status_code == 200:
            data = response.json()
            print(f"\n✔ {data['title']}  —  ${data['price']:.2f}")
        else:
            print(f"\nUnexpected status code: {response.status_code}")
```

**Task 4 — Logging to File**

```python
# Building a Error Log

def log_error(message):
    with open("error_log.txt", "a") as f:
        time = datetime.now()
        f.write(f"[{time}] ERROR: {message}\n")
# Trigger ConnectionError
try:
    requests.get("https://this-host-does-not-exist-xyz.com/api", timeout=5)
except Exception as e:
    log_error(f"ConnectionError - {e}")

# Trigger 404
try:
    res = requests.get(f"{BASE_URL}/999", timeout=5)
    if res.status_code != 200:
        log_error("HTTPError - 404 Not Found for product ID 999")
except Exception as e:
    log_error(str(e))

# Print logs
print("Error Logs:")
with open("error_log.txt", "r") as f:
    print(f.read())
```
